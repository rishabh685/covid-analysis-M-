# READING DATA-SET

```python
india=pd.read_csv("covid_cases.csv")
india
```

# CONVERTING DATATYPE FROM OBJECT TO DATA-TIME

```python
india["Date"]= pd.to_datetime(india["Date"])
india["Date"].dtype
```

# LINE CHART FOR 3-DIFFERENT VARIABLES

```python
tot=px.line(india,"Date",y=["Total Deceased","Total Recovered","Number of cases"],width=1000, height=500)
tot.show()
```

# EXPORTING GRAPH TO PLOTLY SERVER 

```python
py.plot(tot,filename="total_caes")
```

# BAR GRAPH FOR DAILY DECEASED

```python
daily_dead=px.bar(india,"Date","Daily Deceased",color= "Daily Deceased",width=1000, height=500)
daily_dead.show()
```

```python
py.plot(daily_dead,filename="daily death")
```

# ANIMATION

```python
px.area(india,"Date","Number of cases",animation_frame="Daily Confirmed",animation_group="Total Deceased")
```

# AREA GRAPH FOR DAILY CONFIRMED CASES

```python
dcon=px.area(india,"Date","Daily Confirmed",color_discrete_sequence=px.colors.qualitative.Safe)
dcon.show()
```

```python
py.plot(dcon,filename="daily confirmed")
```

# AREA GRAPH FOR DAILY RECOVERED CASES

```python
drec=px.area(india,"Date","Daily Recovered",color_discrete_sequence=["#1CBE4F"])
drec.show()
```

```python
py.plot(drec,filename="daily recovered")
```
